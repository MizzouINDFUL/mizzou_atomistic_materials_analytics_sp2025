```python
#For example, to search for nonionic compounds with registered, IR spectrum, UV-vis spectrum, and phase-change thermodynamic data:

search = nist.Search(NoIon=True, AllowOther=False, AllowExtra=False, cIR=True, cUV=True, cTP=True)
print(search.parameters)
```

```python
#To perform a search for molecules with 10 carbon atoms and any number of hydrogen atoms"
search.find_compounds(f'C10H*', 'formula')
display(search.IDs[1:10])
```

```python
# However, the NIST Chemistry WebBook limits the number of responses to 400. We would benefit from larger datasets for machine learning
print(len(search.IDs))
```

```python
#We want more compounds than that. Let's see how large of a dataset we can build. 
#The below code builds a list of compounds by programmatically adjusting the search terms to capture more compounds
comment this line and run to download a very lot of entries

import pickle
import nistchempy
from tqdm.notebook import tqdm  # Import tqdm notebook for Jupyter-friendly progress bars

# Initialize search and overflow list
overflows = []

# Note: this code takes ~10 minutes to run

# Use tqdm to track progress for the outer loop only
for i in tqdm(range(0, 20), desc="Outer Loop (Carbons)", unit="step"):
    search.find_compounds(f'C{i}H*O*N*S*Cl*', 'formula')
    
    # Check if the limit of 400 compounds has been reached
    if len(search.IDs) == 400:
        for j in range(0, 50):
            search.find_compounds(f'C{i}H{j}O*N*S*Cl*', 'formula')
            if len(search.IDs) != 0:
                if len(search.IDs) == 400:
                    print("2ND LIMIT REACHED")
                else:
                    overflows.extend(search.IDs)
            else:
                continue
    else:
        overflows.extend(search.IDs)

# Print the total number of compounds found
print(f"Total compounds found: {len(overflows)}")

# Save the results to a file
path = './Compounds.pkl'
with open(path, 'wb') as file:
    pickle.dump(overflows, file)
```

```python
import nistchempy
import pickle

# Fetch compound data from NIST using the compound identifier
chem = nistchempy.Compound(ID)

# Extract CAS number, molecular weight, and thermo data references
cas = str(chem.cas_rn)  # CAS number
MW = chem.mol_weight    # Molecular weight
refs = chem.data_refs   # References for data
cTP = refs.get('cTP')   # Thermo data


# Print the relevant data
print(f"CAS Number: {cas}")
print(f"Molecular Weight: {MW} g/mol")

# Check if cTP (thermo data) is available and print it
if cTP:
    print("Thermo data:")
    print(cTP)
else:
    print("No thermo data (constant-pressure heat capacity) available.")
```

```python
#Note that the thermo data is not provided, rather we just have a URL - 
#From the documentation on nistchempy, we read that only the IR and UV data are readily extractable from the NIST database.
#For the other properties, we have to use web scraping to pull the data from the NIST Chemistry WebBook
#Below is a snippet of code to extract one boiling point
# Need to install pubchempy using "pip install pubchempy"

import requests
import bs4
import pandas as pd
import pubchempy

# Extract the web address to the thermo data
address = cTP[0]
web = requests.get(address)
soup = bs4.BeautifulSoup(web.text, 'html.parser')
# Extract the boiling point (if available) from the HTML
BP = None
check = str(soup.find('td'))  # Find the first 'td' tag which should contain the boiling point
if 'boil' in check:
    # Parse the boiling point from the HTML
    BPraw = str(soup.find("td", class_="right-nowrap"))
    BPstr = ''.join([char for char in BPraw if char.isdigit() or char == '.'])  # Extract numeric values
    try:
        BP = float(BPstr)
    except ValueError:
        BP = None

print(BP)
```

```python
# The below code loops over all of the compounds and handles timeout exceptions to extract all of the boiling point data
# Took 128 min to run through 9,324 compounds and got 3,540 hits
comment this line and run to download data for all 3540 compounds

# Import necessary libraries
import pickle
import nistchempy
import requests
import bs4
import pandas as pd
import pubchempy
from tqdm.notebook import tqdm  # For progress bar in Jupyter

compounds = filtered_df['ID']

# Initialize an empty DataFrame to store compound data
CompoundData = pd.DataFrame({"CAS": [], "MW": [], "BP": [], "SMILES": [], "mol3D": []})

# Define output path for saving the resulting DataFrame
output_path = './CompoundData.parquet'

# Loop through each compound, with a progress bar for the outer loop
for i in tqdm(range(len(compounds)), desc="Processing Compounds", unit="compound"):
    try:
        # Fetch compound data from NIST using the compound identifier
        chem = nistchempy.Compound(compounds[i])
        chem.get_3D()
    except Exception:
        # If there's an issue with fetching data, skip to the next compound
        continue

    # Get the CAS number and molecular weight
    cas = str(chem.__getattribute__('cas_rn'))
    MW = chem.__getattribute__('mol_weight')
    threedstruct = chem.mol3D

    # Extract references to phase change data
    refs = chem.__getattribute__('data_refs')
    try:
        # Get the thermo data (cTP) if available
        cTP = refs['cTP']
    except KeyError:
        # If no phase change data is available, skip this compound
        continue

    # Extract the web address to the thermo data
    address = cTP[0]

    # Try to connect to the NIST web page to retrieve boiling point data
    attempt = True
    while attempt:
        try:
            web = requests.get(address)
            soup = bs4.BeautifulSoup(web.text, 'html.parser')
            attempt = False
        except requests.RequestException:
            # Retry connection if there’s a disconnection
            continue

    # Extract the boiling point (if available) from the HTML
    BP = None
    check = str(soup.find('td'))  # Find the first 'td' tag which should contain the boiling point
    if 'boil' in check:
        # Parse the boiling point from the HTML
        BPraw = str(soup.find("td", class_="right-nowrap"))
        BPstr = ''.join([char for char in BPraw if char.isdigit() or char == '.'])  # Extract numeric values
        try:
            BP = float(BPstr)
        except ValueError:
            BP = None

    # If a boiling point is found, proceed to fetch the SMILES string
    if BP is not None:
        pubchem_compound = pubchempy.get_compounds(cas, 'name')
        try:
            # Fetch SMILES string from PubChem
            smile = pubchem_compound[0].canonical_smiles
        except (IndexError, AttributeError):
            smile = None

        # Append the fetched data to the DataFrame
        CompoundData = pd.concat([pd.DataFrame([[cas, MW, BP, smile, threedstruct]], columns=CompoundData.columns), CompoundData], ignore_index=True)

# Display the final DataFrame
display(CompoundData)

# Save the DataFrame to a Parquet file for future use
CompoundData.to_parquet(output_path)
```

```python
#https://github.com/rdkit/rdkit/tree/master/Docs/Notebooks

from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate molecular fingerprints
fingerprints = []
for smiles in CompoundData['SMILES'].dropna():
    mol = Chem.MolFromSmiles(smiles)
    mg = rdFingerprintGenerator.GetMorganGenerator(radius = 2)
    fp = mg.GetFingerprint(mol)
    fingerprints.append(fp)

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
fingerprints_2d = pca.fit_transform(fingerprints)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(fingerprints_2d)

# Visualize clusters in 2D
plt.figure(figsize=(10, 6))
plt.scatter(fingerprints_2d[:, 0], fingerprints_2d[:, 1], c=labels, cmap='viridis')
plt.title('Clusters of Molecular Structures based on Fingerprints')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
